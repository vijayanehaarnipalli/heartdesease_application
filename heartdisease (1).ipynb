```python
#divide into x and y
x= df.drop("target", axis=1)
y= df["target"]
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.7, random_state=99)
```

```python
#train the model
from sklearn.linear_model import LogisticRegression

#load the model
model= LogisticRegression()

#fit the model
train_model= model.fit(x_train, y_train)

#make predictions
y_pred= train_model.predict(x_test)
y_pred
```

```python
#find accuracy
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test, y_pred)
print("accuracy of heart disease model is:" , accuracy)
```

```python
#check loss function
from sklearn.metrics import log_loss
logloss= log_loss(y_test, y_pred)
print("loggloss of logistic regression model is:", logloss)
```

```python
#combined actual vs predicted
comparision= pd.DataFrame({"Actual": y_test, "predicted": y_pred})
comparision
```

```python
#find the confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
cm
```
